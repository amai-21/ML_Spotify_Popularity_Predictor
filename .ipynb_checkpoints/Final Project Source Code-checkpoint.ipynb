# Valuing the U.K. used car by regression models

#### Problem Definition
##### Problem: Predicting the used car price in the U.K.

. Estimating the value of used car in the U.K. by using traditional linear regression model.

. Improving or changing the algorithm, if results from assessment's metrics are not equal or higher than the benchmark on the proposal.

. Data source: https://www.kaggle.com/datasets/adityadesai13/used-car-dataset-ford-and-mercedes

#### Data Description
| Attribute | Description |
| --- | --- |
| model | The specific subcategory of the car |
| year | The manufacturing year of the car |
| price | The value of the car in that dataset's year |
| tramission | The type of transmission system installed in the car, such as manual, automatic, semi-automatic |
| mileage | The total distance traveled by the car, typically measured in miles |
| fuelType | The type of fuel the car uses, such as petrol, diesel, hybrid, and other type  |
| tax | The amount of tax paid for the car, which may vary depending on factors such as engine size, emissions, and local regulations|
| mpg | The fuel efficiency of the car, measured in miles per gallon |
| engineSize | The displacement of the car's engine, usually measured in cubic centimeters (cc) |

## II. Data Cleaning and Feature Engineering

. Replacing the transmission value 'Other' to "Manual" value.

. Encoding object columns.

. Replacing zero values on the engineSize column.

. Calculating the age of car based on the year column.

```python
car_data_copy = car_data.copy()
```

```python
#Replacing the other value with the 'Manual' value
car_data_copy.loc[2458, 'transmission'] = 'Manual'

#Rechecking value counts of transmission
car_data_copy['transmission'].value_counts()
```

```python
#Ecoding the Transmission column
label_econder = preprocessing.LabelEncoder()
car_data_copy['transmission'] = label_econder.fit_transform(car_data_copy['transmission'])

#Rechecking value counts of transmission after encoding
print(car_data_copy['transmission'].value_counts())
```

```python
#Encoding the Fuel Type column
print(car_data['fuelType'].value_counts())

print("\n")

car_data_copy['fuelType'] = label_econder.fit_transform(car_data_copy['fuelType'])
print(car_data_copy['fuelType'].value_counts())
```

```python
#Encoding the Model column
print(car_data['model'].value_counts())

print("\n")

car_data_copy['model'] = label_econder.fit_transform(car_data_copy['model'])
print(car_data_copy['model'].value_counts())
```

```python
#Finding rows with the engine size = 0 
car_data_copy[car_data_copy['engineSize'] == 0]
```

```python
#Finding fitting engine size values
car_list_temp = [2,3,17]
for i in car_list_temp:
    print(car_data_copy[car_data_copy['model'] == i])
```

```python
#Replacing zero values of these engine Size with fitting values
car_data_copy.loc[(car_data_copy['model'] == 2) & (car_data_copy['engineSize'] == 0), 'engineSize'] = 1.0
car_data_copy.loc[(car_data_copy['model'] == 3) & (car_data_copy['engineSize'] == 0), 'engineSize'] = 1.8
car_data_copy.loc[(car_data_copy['model'] == 17) & (car_data_copy['engineSize'] == 0), 'engineSize'] = 1.5

car_data_copy['engineSize'].value_counts()
```

```python
#Calculating the age of car, dropping the year column and making a new column called 'age'
car_data_copy['age'] = 2024 - car_data_copy['year']
car_data_copy = car_data_copy.drop(columns=['year'])
car_data_copy.head()
```

## IV. Data preparation for training and testing

```python
#Splitting columns of the dataset into independent variables and the dependent variable
ind_cols = ['model','transmission','mileage','fuelType','tax','mpg','engineSize','age']
x = car_data_copy.loc[:, ind_cols]
y = car_data_copy['price']
```

```python
#Splitting the dataset into the trainning dataset and the testing dataset by the 80/20 proportion
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12345)
```

## V. Building the linear regression model

```python
#Initializing the linear model
linear_regression = linear_model.LinearRegression()
```

```python
#Fitting x_train and y_train into the linear regression model
lr = linear_regression.fit(x_train, y_train)
```

```python
#Making the price prediction from the model
reg_pred_train = lr.predict(x_train)
reg_pred_test = lr.predict(x_test)
```

```python
#Showing the actual price, predicted price, and the difference
model_df = pd.DataFrame({'Actual Price': y_test, 'Predicting Price': reg_pred_test})
model_df.head(10)
```

## VI. Assessment the model

. Calculating R-squared, MSE, and RMSE

. Showing the correlation graph of predict price and actual price


```python
#Calculating R2, Mean Absolute Error, and Mean Squared Error
mse_train_1 = round(mean_squared_error(y_train, reg_pred_train),2)
rmse_train_1 = round(np.sqrt(mse_train_1),2)
r2_train_1 = round(r2_score(y_train, reg_pred_train),2)
mse_test_1 = round(mean_squared_error(y_test, reg_pred_test),2)
rmse_test_1 = round(np.sqrt(mse_test_1),2)
r2_test_1 = round(r2_score(y_test, reg_pred_test),2)

scores_1 = []
scores_1.append([mse_train_1, rmse_train_1, r2_train_1])
scores_1.append([mse_test_1, rmse_test_1, r2_test_1])

scores_1 = pd.DataFrame(scores_1, columns=['MSE', 'RMSE', 'R2'], index=['Training', 'Testing'])
scores_1
```

```python
#Graphing the model for checking the difference between actual price and predicted price
#Plotting the scatter plot of the actual vs. predicted values
plt.scatter(y_test, reg_pred_test, color='orange')
plt.title('Actual vs. Predicted Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')

#Plotting the regression line
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='blue')  # Diagonal line
plt.show()
```

## Tuning up the model by using the KNN

. Categorizing the price of cars into 3 type: Cheap (<= 10000), Suitable (<= 20000), Expensive (> 20000)

. Using KNN algorithm by the number of default neigbors (5)

. Metrics of Assessment F1-Score, Precision, Accuracy.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
car_data_copy['priceStatus'] = np.where(car_data_copy.price <= 10000, 'Cheap', np.where(car_data_copy.price <= 20000, 'Suitable', 'Expensive'))
car_data_copy.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
car_data_copy['priceStatus'] = label_encoder.fit_transform(car_data_copy['priceStatus'])
new_data = car_data_copy.drop(columns=['price'])
```

```python
X = new_data.drop(['priceStatus'], axis = 1)
Y = new_data.priceStatus
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

scaler = preprocessing.StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y = Y_train)
```

```python
train_preds = knn.predict(X=X_train_scaled)
test_preds = knn.predict(X=X_test_scaled)
```

```python
scores_c = {}
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores_c['training'] = [round(accuracy_score(Y_train, train_preds), 5),\
                     round(precision_score(Y_train, train_preds, average='weighted'), 5),\
                     round(recall_score(Y_train, train_preds, average='weighted'), 5),\
                     round(f1_score(Y_train, train_preds, average='weighted'), 5)]
scores_c['testing'] = [round(accuracy_score(Y_test, test_preds), 5),\
                     round(precision_score(Y_test, test_preds, average='weighted'), 5),\
                     round(recall_score(Y_test, test_preds, average='weighted'), 5),\
                     round(f1_score(Y_test, test_preds, average='weighted'), 5)]
scores_df = pd.DataFrame(scores_c, index=metrics)
scores_df
```

## VII. Tuning up the model by using the Gradient Boosting Regression

. A machine learning technique that combines multiple weaker models sequentially to build a strong predictive model, similar to assembling a team of players.

. It uses the direction of steepest ascent in a function, like climbing a hill, to adjust the model iteratively. By taking small steps towards the optimal solution, the prediction errors are gradually reduced.

. It uses decision trees, particularly regression trees, as the base learner for regression tasks. Each tree predicts the residuals from the previous trees, and by adding them together, we iteratively reduce the errors in our predictions. This leads to a more accurate model, with each tree focusing on the remaining errors to improve upon.

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
GB = GradientBoostingRegressor(random_state=0)
GB.fit(x_train,y_train)
```

```python
GB_pred_train = GB.predict(x_train)
GB_pred_test = GB.predict(x_test)

#Calculating R2, Mean Absolute Error, and Mean Squared Error
mse_train_2 = round(mean_squared_error(y_train, GB_pred_train),2)
rmse_train_2 = round(np.sqrt(mse_train_2),2)
r2_train_2 = round(r2_score(y_train, GB_pred_train),2)
mse_test_2 = round(mean_squared_error(y_test, GB_pred_test),2)
rmse_test_2 = round(np.sqrt(mse_test_2),2)
r2_test_2 = round(r2_score(y_test, GB_pred_test),2)

scores_2 = []
scores_2.append([mse_train_2, rmse_train_2, r2_train_2])
scores_2.append([mse_test_2, rmse_test_2, r2_test_2])
scores_2 = pd.DataFrame(scores_2, columns=['MSE', 'RMSE', 'R2'], index=['Training', 'Testing'])
scores_2
```

```python
#Graphing the model for checking the difference between actual price and predicted price
#Plotting the scatter plot of the actual vs. predicted values
plt.scatter(y_test, GB_pred_test, color='orange')
plt.title('Actual vs. Predicted Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')

#Plotting the regression line
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='blue')  # Diagonal line
plt.show()
```

```python
# Plot outputs 
plt.scatter(x_test.mileage, y_test,  color='black') 
plt.title('Test Data') 
plt.xlabel('Mileage') 
plt.ylabel('Price') 
plt.xticks(()) 
plt.yticks(()) 
plt.plot(x_test.mileage, GB_pred_test, color='red',linewidth=1) 
plt.show() 
```
