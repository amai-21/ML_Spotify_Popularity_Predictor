```python
#Data preprocessing; Variables with distinct values and high cardinality are not helpful. Therefore, we'll drop those
#variables, i.e. track, artist, and uri.
data.drop(['track', 'artist', 'uri'], axis = 1, inplace=True)
```

```python
#Split the training data into input and output sets:
y = data['target']
X= data.drop('target', axis = 1)
```

```python
#Splitting the data for Training and Testing
#70% of the data is used for training and 30% is used for testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, shuffle = True, random_state = 1)

#Scale:
X_train = pd.DataFrame(preprocessing.robust_scale(X_train), index=X_train.index)
X_test = pd.DataFrame(preprocessing.robust_scale(X_test), index=X_test.index)
```

```python
#Use Scikit-Learn's GridSearchCV() to tune the parameters:
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
from sklearn.model_selection import GridSearchCV
# Define the hyperparameter grid:

#n_estimators: 150
#max_depth: 300
#min_samples_split: 250
#min_samples_leaf: 80
#max_features: ['auto','sqrt','log2']
#max_samples: 300

param_grid = {
      'n_estimators': np.arange(10,200,10),
    'min_samples_leaf': np.arange(1,100,10),
    'max_features': ['auto', 'sqrt', 'log2']
}


#Create based model:
rf = RandomForestClassifier()
```

```python
#Instantiate GridSearchCV
grid_search = GridSearchCV(estimator = rf, param_grid=param_grid, cv=3, scoring='accuracy')
```

```python
# Fit GridSearchCV to the training data:
grid_search.fit(X_train, y_train)
#570 fits failed out of a total of 1710
```

```python
# Access the Best hyperparameters found by GridSearchCV
best_params = grid_search.best_params_

# Access the best model:
best_model = grid_search.best_estimator_

```

```python
print('Best parameters:', best_params)
print('Best Model:', best_model)
```
